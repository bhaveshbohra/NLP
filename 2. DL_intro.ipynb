## **Implementing the core deep learning models - MLPs, CNNs, and RNNs**

We'll be using three basic deep learning models, for the sake of introduction or revision in this lecture. They are:

* RNNs: Recurrent neural networks

## **The difference between MLPs, CNNs, and RNNs**


Multilayer perceptrons or MLPs are a fully-connected network. You'll often find
them referred to as either deep feedforward networks or feedforward neural
networks in some literature. MLPs are common in simple logistic
and linear regression problems. 

However, MLPs are not optimal for processing
sequential and multi-dimensional data patterns. By design, MLPs struggle to
remember patterns in sequential data and requires a substantial number of
parameters to process multi-dimensional data.

For sequential data input, RNNs are popular because the internal design allows
the network to discover dependency in the history of data that is useful for
prediction. 

For multi-dimensional data like images and videos, a CNN excels
in extracting feature maps for classification, segmentation, generation, and other
purposes. 

In some cases, a CNN in the form of a 1D convolution is also used for
networks with sequential input data. 

However, in most deep learning models, MLPs, RNNs, and CNNs are combined to make the most out of each network.

## **Other actors of the drama**

MLPs, RNNs, and CNNs do not complete the whole picture of deep networks. There is a need to identify an objective or loss function, an optimizer, and a regularizer.

The goal is to reduce the loss function value during training since it is a good guide that a model is learning. 

To minimize this value, the model employs an optimizer.
This is an algorithm that determines how weights and biases should be adjusted
at each training step. 

A trained model must work not only on the training data but
also on a test or even on unforeseen input data. The role of the regularizer is to
ensure that the trained model generalizes to new data.

# <font color='brown'>**Building a model using MLPs and Keras**</font>

image.png

After data preparation, next step is building the model. 

In Keras, an MLP layer is referred to as **Dense**, which stands for the **densely connected layer**.

The proposed model is made of three MLP layers. Both the first and second MLP layers are identical in nature with 256 units each, followed by relu activation and dropout. 

256 units are chosen since 128, 512 and 1,024 units have lower performance metrics. At 128 units, the network converges quickly, but has a lower test accuracy. The added number units for 512 or 1,024 does not increase the test accuracy significantly.

The number of units, which is a **hyperparameter**, controls the capacity of the network.

The capacity is a measure of the complexity of the function that the network can
approximate. For example, for polynomials, the degree is the hyperparameter.
As the degree increases, the capacity of the function also increases.

As shown in the following model, the classifier model is implemented using a sequential model API of Keras.

```
# network parameters
batch_size = 128 #  Epoch contains 60,000 samples ; no of batches = 60000/128 
hidden_units = 256
dropout = 0.45
```

```
# model is a 3-layer MLP with ReLU and dropout after each layer
model = Sequential()
# model.add(Dense(hidden_units, input_dim=input_size))
model.add(Dense(256, input_dim=784))
model.add(Activation('relu'))
model.add(Dropout(dropout))
model.add(Dense(hidden_units))
model.add(Activation('relu'))
model.add(Dropout(dropout))
model.add(Dense(num_labels))
# this is the output for one-hot vector
model.add(Activation('softmax'))
model.summary()
```

## **Activation Functions**

Dense layer is a linear operation, a sequence of Dense layers can only approximate a linear function. 

The problem is that the **MNIST digit classification is non-linear process**. 

Inserting a non-linear activation between Dense layers will enable MLPs to model non-linear mappings. 

Relu or **Rectified Linear Unit (ReLU)** is a simple non-linear function. It's like a filter that allows positive inputs to pass through unchanged while clamping everything else to zero. 

Mathematically, relu is expressed as: 

$$
    y = max(0,x) 
$$



**Leaky ReLU**

Leaky Rectified Linear Unit, or Leaky ReLU, is a type of activation function based on a ReLU, but it has a small slope for negative values instead of a flat slope. 

The slope coefficient is determined before training, i.e. it is not learnt during training. 

This type of activation function is popular in tasks where we we may suffer from sparse gradients, for example training generative adversarial networks.

$$
    y = ax
$$

In such a case, the network fails when subjected to test data. This is a classic case when network **fails to generalize**. 

**To avoid this tendency, the model uses a regularizing layer or function.**

### **Dropout**

One of the most common regularization tehcnique followed in deep learning is Dropout.

The Dropout layer is a **mask** that nullifies the contribution of some neurons towards the next layer and leaves others unchanged. 

We can apply a Dropout layer to the **input vector**, in which case it nullifies some of its features; 

We can also apply Dropout to a **hidden layer**, in which case it nullifies some hidden neurons.

![](https://www.baeldung.com/wp-content/uploads/sites/4/2020/05/2-1-2048x745-1-1536x559.jpg)

Given a dropout rate (here, it is set to dropout=0.45), the Dropout layer randomly removes that fraction of units from participating in the next layer. 

As the first layer has 256 units, after $\text{dropout}=0.45$ is applied, only $(1 - 0.45) \times 256 = 140$ units from layer 1 participate in layer 2.

The Dropout layer makes neural networks robust to unforeseen input data because
the network is trained to predict correctly, even if some units are missing. 

Dropout is not used in the output layer and it is only active during training. 

### **L1, L2 regularization**

L1 and L2 are the most common types of regularization. These update the general cost function by adding another term known as the regularization term.

$$
\text{Cost function = Loss (eg. binary cross entropy) + Regularization term}
$$

Due to the addition of this regularization term, the values of weight matrices decrease because it assumes that a neural network with smaller weight matrices leads to simpler models. Therefore, it will also reduce overfitting to quite an extent.

However, this regularization term differs in L1 and L2.

### **L2**

It adds an L2 penalty which is equal to the square of the magnitude of coefficients. Ridge regression and SVM implement this method.

$$
\text{Cost function = Loss (eg. binary cross entropy)} + \lambda \Sigma \|w\|^2 
$$

Here, $\lambda$ is the regularization parameter. It is the hyperparameter whose value is optimized for better results. 

L2 regularization is also known as **weight decay** as it forces the weights to decay towards zero (but not exactly zero).


### **L1**
It adds an L1 penalty that is equal to the absolute value of the magnitude of coefficient, or simply restricting the size of coefficients. Lasso regression implements this method.
$$
\text{Cost function = Loss (eg. binary cross entropy)} + \lambda \Sigma \|w\|
$$

In this, we penalize the absolute value of the weights. 

Unlike L2, the weights may be reduced to zero here. Hence, it is very useful when we are trying to compress our model. Otherwise, we usually prefer L2 over it.
    from keras import regularizers

    # To add L2 regularizer in a layer
    model.add(Dense(64, input_dim=64, kernel_regularizer=regularizers.l2(0.01))

    # To add L1 regularizer in a layer
    model.add(Dense(64, input_dim=64, kernel_regularizer=regularizers.l1(0.0001))


No additional layer is added if l1 or l2 regularization is used. 

The regularization is imposed in the Dense layer internally.

## **Output activation and loss function**

The output layer has 10 units followed by softmax activation. 

The 10 units correspond to the 10 possible labels, classes or categories. 

The softmax activation can be expressed mathematically as shown in the following equation:

$$
  softmax(x_i)=\frac{e^{x_i}}{\sum_{j=0}^{N-1} e^{x_j}}
$$

The equation is applied to all $N = 10$ outputs, $x_i$ for $i = 0, 1 ‚Ä¶ 9$ for the final prediction.

It squashes the outputs into probabilities by normalizing the prediction. 

Here, each predicted output is a probability that the index is the correct label of the given input image. 

The sum of all the probabilities for all outputs is 1.0.

For example, when the softmax layer generates a prediction, it will
be a 10-dim 1D tensor that may look like the following output:

The prediction output tensor suggests that the input image is going to be 7 given that its index has the highest probability.

# <font color='brown'>**Training neural networks**</font>

We have seen how neural networks can map inputs onto determined outputs, depending on fixed weights. 

Once the architecture of the neural network has been defined and includes the feed forward network, the number of hidden layers, the number of neurons per layer, and the activation function, we'll need to set the weights, which, in turn, will define the internal states for each neuron in the network. 



Every neural network is an approximation of a function, so each neural network will not be equal to the desired function, but instead will differ by some value called error. 

During training, the aim is to minimize this error. Since the error is a function of the weights of the network, we want to minimize the error with respect to the weights. 

The error function is a function of many weights and, therefore, a function of many variables. 

Mathematically, the set of points where this function is zero represents a hypersurface, and to find a minimum on this surface, we want to pick a point and then follow a curve in the direction of the
minimum.

**Note**

> We should note that a neural network and its training are two separate
things. This means we can adjust the weights of the network in some way
other than gradient descent and backpropagation, but this is the most
popular and efficient way to do so and is, ostensibly, the only way that is
currently used in practice.

## <font color='blue'>**Backpropagation Algorithm**</font>

### **Notations**
> <font color='red'>**Source:**</font> [Artificial Intelligence](https://www.edx.org/course/artificial-intelligence-ai)


If the current neuron (neuron under consideration) is $j$
* $x_{ij}$: input from $i^{th}$ neuron to $j^{th}$ neuron
* $w_{ij}$: weight associated with the edge from $i^{th}$ neuron to $j^{th}$ neuron
* $z_{j}=\sum w_{ij}x_{ij}$: weighted sum of inputs from $i^{th}$ neuron to $j^{th}$ neuron (legit)
* $o_j=g(z_j)$: output computed by $j^{th}$ neuron
* $g=\frac{1}{1+e^{-z_j}}$ is the sigmoid function
* $outputs$: set of neurons in the output layer
* $Succ(j)$: set of neurons which are getting output of neuron $j$ as one of inputs.

Output of a neuron can be given as:
\begin{align}
    o_j &= \frac{1}{1+e^{-z_j}} \tag{1}\\
    o_j &= \frac{1}{1+e^{-\sum w_{ij}x_{ij}}} \\
\end{align}

### **Error Computation**

Suppose we have $k$ outputs.

Take an example $e$ represented as $(x,y)$ where $x$ is input and $y$ is the output.

The error on e, summed over all $k$ outputs neurons is:
\begin{equation}
    \textrm{SSE}=E(w)=\frac{1}{2} \sum_k \left( y_k-o_k \right)^2  \tag{2}
\end{equation}
Where SSE means Sum of Squared Errors


from $(1)$ and $(2)$
\begin{equation}
    E(w)=\frac{1}{2} \sum_k \left[ y_k- \left( \frac{1}{1+e^{-\sum w_{jk}x_{jk}}} \right) \right]^2 \tag{3}
\end{equation}



### **Gradient descent**

Training of neural network depends on updation of weight values. 

In a generic format, updation of a weight can be given as:
\begin{equation}
    w'_{ij}=w_{ij}+\Delta w_{ij} \tag{4}
\end{equation}


We can utilize Gradient Descent algorithm to calculate $\Delta w_{ij}$. Since $E$ is a function of $w$ (see $(3)$), we have the gradient term as:
\begin{equation}
    \Delta w_{ij}=-\alpha \frac{\partial E}{\partial w_{ij}} \tag{5}
\end{equation}

Using gradient descent ensures that with every upation to the old value of $w_{ij}$ we will successively move towards a lower value of error $E(w_{ij})$.

Therefore, with $(4)$ and $(5)$ we get:
\begin{equation}
    w'_{ij}=w_{ij}-\alpha \frac{\partial E}{\partial w_{ij}} \tag{6}
\end{equation}

Now our task is to simplify the gradient term of $(6)$ to obtain an expressions in terms of known values in a neural network.

### **The Chain Rule**

The chain rule is a mathematical theorem that lets us compute derivatives of composite functions. 

Deep learning models are, mathematically, composite functions, and reasoning about their derivatives is essential to training them.

Using chain rule, we can write:

\begin{equation}
    \frac{\partial E}{\partial w_{ij}} = \frac{\partial E}{\partial z_j}\frac{\partial z_j}{\partial w_{ij}} \tag{7} 
\end{equation}

As we know:
\begin{align}
    z_j &= \sum w_{ij}x_{ij} = w_{0j}x_{0j}+w_{1j}x_{1j}+w_{2j}x_{2j}+...+w_{ij}x_{ij}+...+w_{nj}x_{nj} \\
    \because \frac{\partial z_j}{\partial w_{ij}} &= \frac{\partial(w_{0j}x_{0j}+w_{1j}x_{1j}+w_{2j}x_{2j}+...+w_{ij}x_{ij}+...+w_{nj}x_{nj})}{\partial w_{ij}} \\
    \Rightarrow \frac{\partial z_j}{\partial w_{ij}} &= x_{ij} \tag{8}
\end{align}





From $(7)$ and $(8)$, 

\begin{equation}
    \frac{\partial E}{\partial w_{ij}} = \frac{\partial E}{\partial z_j}x_{ij} \tag{9} 
\end{equation}

Thus from $(5)$ and $(9)$ we have

\begin{equation}
    \Delta w_{ij}=-\alpha \frac{\partial E}{\partial z_j}x_{ij} \tag{10}
\end{equation}

We now consider two case on the basis of two types of neurons:

1. Neuron $j$ is an <font color='blue'>**output neuron**</font>
2. Neuron $j$ is a <font color='blue'>**hidden neuron**</font>

#### <font color='red'>**Case 1. Neuron `j` is an Output Neuron**</font>

\begin{equation}
    \frac{\partial E}{\partial z_j} = \frac{\partial E}{\partial o_j} \frac{\partial o_j}{\partial z_j} \tag{11}
\end{equation}

We divide right hand side of the above equation in two parts:
\begin{equation}
    \frac{\partial E}{\partial o_j} 
\end{equation}

and
\begin{equation}
    \frac{\partial o_j}{\partial z_j} 
\end{equation}

<font color='blue'>**Solving Part-1:**</font>

\begin{equation}
    \frac{\partial E}{\partial o_j} = \frac{\partial}{\partial o_j} \left[ \frac{1}{2} \sum_k \left( y_k-o_k \right)^2 \right] \\
\end{equation}

Since we are differentiating w.r.t. $o_j$, only the output neuron $j$ will be useful while output of rest of the neurons will differentiate to $0$. 

Therefore,
\begin{align}
    \frac{\partial E}{\partial o_j} &= \frac{\partial}{\partial o_j} \left[ \frac{1}{2}  \left( y_j-o_j \right)^2 \right] \\
    \frac{\partial E}{\partial o_j} &= \frac{1}{2} 2 (y_j - o_j) \frac{\partial (y_j-o_j)}{\partial o_j} \\
    \frac{\partial E}{\partial o_j} &= (y_j - o_j) (-1) \\
    \frac{\partial E}{\partial o_j} &= -(y_j - o_j) \tag{12}
\end{align}


<font color='blue'>**Solving Part-2:**</font>

From $(1)$:
\begin{align}
    \frac{\partial o_j}{\partial z_j} &= \frac{\partial}{\partial z_j} \left( \frac{1}{1+e^{-z_j}} \right) \\
    &= \frac{\partial}{\partial z_j} \left( 1+e^{-z_j} \right)^{-1} \\
    &= - \left( 1+e^{-z_j} \right)^{-2}\frac{\partial}{\partial z_j} \left( 1+e^{-z_j} \right) \\
    &= - \left( 1+e^{-z_j} \right)^{-2}\frac{\partial}{\partial z_j} \left( e^{-z_j} \right) \\
    &= \left( 1+e^{-z_j} \right)^{-2} ( e^{-z_j} ) \\
    &= \frac{e^{-z_j}}{(1+e^{-z_j})^2} \\
    &= \frac{1 + e^{-z_j} -1}{(1+e^{-z_j})^2} \\
    &= \frac{1 + e^{-z_j}}{(1+e^{-z_j})^2} - \frac{1}{(1+e^{-z_j})^2} \\
    &= \frac{1}{(1+e^{-z_j})} - \frac{1}{(1+e^{-z_j})^2} \\
    &= \left( \frac{1}{1+e^{-z_j}} \right)\left(1 - \frac{1}{1+e^{-z_j}} \right) \\
\end{align}

From $(1)$ again,
\begin{align}
    \frac{\partial o_j}{\partial z_j} &= o_j(1 - o_j) \tag{13}\\
\end{align}

Finally, from $(11),(12)$ and $(13)$ we have:

\begin{equation}
    \frac{\partial E}{\partial z_j} = -(y_j - o_j) o_j(1 - o_j) 
\end{equation}

Combining $(10)$ we get:
\begin{equation}
    \Delta w_{ij} = \alpha (y_j - o_j) o_j(1 - o_j)x_{ij}  
\end{equation}

Suppose 
\begin{equation}
    \delta_j = -\frac{\partial E}{\partial z_j} = (y_j - o_j) o_j(1 - o_j)  \tag{14}
\end{equation}

We get,
\begin{equation}
    \Delta w_{ij} = \alpha \delta_j x_{ij}  \tag{15}
\end{equation}

#### <font color='red'>**Case 2. Neuron `j` is a Hidden Neuron**</font>

\begin{equation}
    \frac{\partial E}{\partial z_j} = \sum_{k \in succ\{j\}} \frac{\partial E}{\partial z_k} \frac{\partial z_k}{\partial z_j}
\end{equation}

from $(14)$,
\begin{equation}
    \frac{\partial E}{\partial z_j} = \sum_{k \in succ\{j\}} -\delta_k \frac{\partial z_k}{\partial z_j}  \tag{16}
\end{equation}

Further, using chain rule:
\begin{equation}
    \frac{\partial z_k}{\partial z_j} = \frac{\partial z_k}{\partial o_j} \frac{\partial o_j}{\partial z_j} \tag{17}
\end{equation}

<font color='blue'>**Note**</font>: The inputs to a $k^{th}$ neuron is the outputs produced by all $j^{th}$ neurons. 

So,
\begin{equation}
    z_k = \sum_j w_{jk}o_{jk} = w_{0k}o_{0k}+w_{1k}o_{1k}+...+w_{jk}o_{jk}+...
\end{equation}

Therefore we have,
\begin{equation}
    \frac{\partial z_k}{\partial o_j} = \frac{\partial }{\partial o_j} [w_{0k}o_{0k}+w_{1k}o_{1k}+...+w_{jk}o_{jk}+...] = w_{jk}
\end{equation}

Taking it to $(17)$, we get
\begin{equation}
    \frac{\partial z_k}{\partial z_j} = w_{jk} \frac{\partial o_j}{\partial z_j} \tag{18}
\end{equation}

Also from $(13)$, we know that 
\begin{align}
    \frac{\partial o_j}{\partial z_j} &= o_j(1 - o_j) 
\end{align}

Thus combining $(13), (18)$ with $(16)$ we get:
\begin{align}
    \frac{\partial E}{\partial z_j} &= \sum_{k \in succ\{j\}} -\delta_k w_{jk} o_j(1 - o_j)  \\
    \frac{\partial E}{\partial z_j} &= -o_j(1 - o_j)\sum_{k \in succ\{j\}} \delta_k w_{jk}   \\
    -\frac{\partial E}{\partial z_j} &= o_j(1 - o_j)\sum_{k \in succ\{j\}} \delta_k w_{jk}   \tag{19}
\end{align}

As taken in $(14)$ before:
\begin{equation}
    \delta_j = -\frac{\partial E}{\partial z_j} = o_j(1 - o_j)\sum_{k \in succ\{j\}} \delta_k w_{jk} \tag{20}
\end{equation}



### <font color='black'>**Summarizing**</font>

As mentioned before in $(4)$ we get:
\begin{align}
    w'_{ij} &= w_{ij}+\Delta w_{ij}
\end{align}

where from $(15)$ we get:
\begin{align}
    \Delta w_{ij} &= \alpha \delta_j x_{ij}
\end{align}

where in case of $j$ being an output neuron, from $(14)$ we get,
\begin{align}
    \delta_j &= (y_j - o_j) o_j(1 - o_j)
\end{align}


and when $j$ is a hidden neuron, from $(20)$ we get,
\begin{align}
    \delta_j &= o_j(1 - o_j)\sum_{k \in succ\{j\}} \delta_k w_{jk}
\end{align}

### <font color='black'>**Final Algorithm**</font>


**Input:** training examples $(x,y)$, learning rate $\alpha$ (e.g., $\alpha = 0.1$), $n_i, n_h$ and $n_o$.

**Output:** a neural network with one input layer, one hidden layer and one output layer with $n_i, n_h$ and $n_o$ number of neurons respectively and all its weights.

<ol>
    <li>Create_feedforward_network ($n_i, n_h$,$n_o$)</li>
    <li>Initialize all weights to small random numbers (e.g., in $[-0.2,0.2]$</li>
    <li>Repeat until convergence</li>
    For each training sample $(x,y)$
        <ol>
            <li seq="a"><font color='blue'>Feedforward:</font> Propagate example $x$ through the network and compute the output $o_j$ from every neuron.</li>
            <li  seq="b"><font color='blue'>Propagate backwards:</font> Propagate the erros backward. 
                </br>Case 1: For each output neuron $k$, calculate its error <font color='red'>$$\\ \delta_k = o_k(1-o_k)(y_k-o_k)$$</font>
                </br>Case 2: For each hidden neuron $h$, calculate its error <font color='red'>$$\\ \delta_h = o_h(1-o_h)\sum_{k \in Succ(h)}w_{hk}\delta_k$$</font>
            </li>
            <li  seq="c">Update each weight as <font color='red'>$$\\w_{ij} \leftarrow w_{ij} + \alpha \delta_j x_{ij}$$</font></li>
    </ol>

## **Optimization**

With optimization, the objective is to minimize the loss function. 

The idea is that **if the loss is reduced to an acceptable level, the model has indirectly learned the function mapping input to output.**

Performance metrics are used to determine if a model has learned the underlying data distribution. The default metric in Keras is loss. 

During training, validation, and testing, other metrics such as accuracy
can also be included. 

Accuracy is the percent, or fraction, of correct predictions based on ground truth. 

In deep learning, there are many other performance metrics. However, it depends on the target application of the model. 

In literature, performance metrics of the trained model on the test dataset is reported for comparison to other deep learning models.

The most commonly used optimizers are; 
* Stochastic Gradient Descent (SGD), 
* Adaptive Moments (Adam), and
* Root Mean Squared Propagation (RMSprop). 

Each optimizer has tunable parameters like learning rate, momentum, and decay. 

Adam and RMSprop are variations of SGD with adaptive learning rates. 

In the proposed classifier network, Adam is used since it has the highest test accuracy.

SGD is considered the most fundamental optimizer. It's a simpler version of the
gradient descent in calculus. 

In gradient descent (GD), tracing the curve of a function downhill finds the minimum value, much like walking downhill in a valley or opposite the gradient until the bottom is reached.
